```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#invocando o dataframe a partir do arquivo csv
df = pd.read_csv('continuous_factory_process.csv', sep=',', encoding='utf-8')


#selecionando colunas relacionadas a operação da máquina 1
filter = ['AmbientConditions.AmbientHumidity.U.Actual',
 'AmbientConditions.AmbientTemperature.U.Actual',
 'Machine1.RawMaterial.Property1',
 'Machine1.RawMaterial.Property2',
 'Machine1.RawMaterial.Property3',
 'Machine1.RawMaterial.Property4',
 'Machine1.RawMaterialFeederParameter.U.Actual',
 'Machine1.Zone1Temperature.C.Actual',
 'Machine1.Zone2Temperature.C.Actual',
 'Machine1.MotorAmperage.U.Actual',
 'Machine1.MotorRPM.C.Actual',
 'Machine1.MaterialPressure.U.Actual',
 'Machine1.MaterialTemperature.U.Actual',
 'Machine1.ExitZoneTemperature.C.Actual']

#criando dataframe filtrando
df_filter = df[filter]

#invocando função describe com Transpose para melhorar visualização
df_filter.describe().T

```

```python
# Configurando o tamanho da figura e fonte
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(90,50))

sns.pairplot(df_filter, corner=True, diag_kind="kde",height=3.0)
```
